```python
import pandas as pd

import numpy as np

import glob

import openpyxl

import warnings
warnings.simplefilter("ignore")
```

```python
#getting all the required files from the location

path = r'C:/Users/agheorghitoiu/Downloads/Google Analytics'

all_files = glob.glob(path + '/*.xlsx')

li = []
```

```python
#concatening the files by the specified sheet name

for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0, sheet_name='Dataset1')
    li.append(df)
    
frame =pd.concat(li, axis=0, ignore_index=True)
```

```python
#checking the newly resulted dataframe

frame
```

```python
#getting more info about the dataframe with the .describe() method

frame.describe()


```

```python
#creating a new column containing labels for the URLs
#after checking the dataframe, the expressions below contain most of the groups
#a condition list was created and also a choice list, to match a condition with a choice, e.g. trados with Language Technology, tridion with Content Management

condlist = [
    frame['Page'].str.contains('.*trados', na=False),
    frame['Page'].str.contains('.*translation_management', na=False),
    frame['Page'].str.contains('.*translationproductivity', na=False),
    frame['Page'].str.contains('.*linguistic-ai', na=False), 
    frame['Page'].str.contains('.*tridion', na=False),
    frame['Page'].str.contains('.*contenta', na=False),
]

choicelist = ['Language Technology', 'Language Technology', 'Language Technology', 'Artficial Intelligence', 'Content Management', 'Content Management']

frame['Product_Group'] = np.select(condlist, choicelist)

frame
```

```python
#filtering results only from community.rws.com

views = frame[frame['Page'].str.contains('.*community.rws.com', na=False)]

views
```

```python
#creating a new dataframe with an additional column with the hardcoded month and year

monthly_views = views.assign(month_year='Jan 2023')

monthly_views
```

```python
#creating an Excel file based on the newly resulted dataframe

writer = pd.ExcelWriter('Google_Analytics_Views_Jan2023.xlsx',
                       engine='xlsxwriter',
                       engine_kwargs={'options': {'strings_to_urls':False}}
                       )
monthly_views.to_excel(writer)

writer.close()
```
